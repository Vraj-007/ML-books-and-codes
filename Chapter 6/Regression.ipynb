```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-10, 10, 300)

y = np.zeros(X.shape)

for i in range(len(X)):
    noise = np.random.rand(1, 1)*20
    y[i] = X[i]**2 + noise

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

plt.scatter(X, y)
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(max_depth=3)

model.fit(X, y)
```
